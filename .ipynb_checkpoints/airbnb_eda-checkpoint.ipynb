```python
def missing_values(df):
	missing_percentage = df.isnull().sum()*100/len(df)
	plt.figure(figsize=(5, 15))
	missing_percentage.plot(kind='barh')
	plt.xticks(rotation=90, fontsize=10)
	plt.yticks(fontsize=5)
	plt.xlabel("Missing Percentage", fontsize=14)
	plt.show()

missing_values(df)
```

```python
#Unique returns values in the dataset without duplicates and also enables us to see missing values,if there are.
def unique_missing_values(df):
  #Checking existence of null values.
  null_sum = [df[column].isnull().sum() for column in list(df.columns.values)]
  null_sum_dict = dict(zip(list(df.columns.values),null_sum))
  print('NULL VALUE COUNTS PER FEATURE')
  print(null_sum_dict)
  #Unique values that are in features.
  for column in df.columns:
    print(column.upper())
    print(df["{}".format(column)].unique())
    
unique_missing_values(df)
```

# Preprocess

```python
def drop_missing_columns(df,percentage):
    missing_percentage = df.isnull().sum()*100/len(df)
    features_left = missing_percentage[missing_percentage < percentage].index
    return df[features_left]
df = drop_missing_columns(df,10)
df
```

```python
#Converts features that have letters into categorical.
def convert_numerical(df):
    possible_categorical_columns = numerical_features = df.dtypes[df.dtypes == "object"].index.to_list()
    letter_pattern = re.compile(r'[A-z]')
    special_pattern = re.compile('-')
    
    for column in possible_numerical_columns:
        if not letter_pattern.match(str(df[column].values[0])) and special_pattern.match(str(df[column].values[0])):
            df[column] = pd.to_numeric(df[column])
        
    return df
convert_numerical(df)
```
